**Data Preprocessing**

```python
#Convert categorical into numerical values
data_pre=pd.get_dummies(data,columns=['Fuel_Type','Selling_type','Transmission'])
```

```python
data_pre.info()
```

```python
data_pre.head()
```

```python
from sklearn.preprocessing import StandardScaler

Scalar=StandardScaler()
data_pre[['Year','Selling_Price','Present_Price','Driven_kms']]=Scalar.fit_transform(data_pre[['Year','Selling_Price','Present_Price','Driven_kms']])
```

```python
data_pre.head()
```

```python
#convert boolean columns into integer
bool_colum=data_pre.select_dtypes(include='bool').columns
data_pre[bool_colum]=data_pre[bool_colum].astype(int)
```

```python
data_pre.head()
```

```python
data_pre.info()
```

**Feature selection**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_pre_num=data_pre.select_dtypes(include='number')
sns.heatmap(data_pre_num.corr(), annot=True,cmap='coolwarm')
```

```python
#Splitting into Target and Feature Variable
from sklearn.model_selection import train_test_split

X=data_pre.drop(['Selling_Price','Car_Name'],axis=1)
Y=data_pre['Selling_Price']
#Splitting into testing and training Values
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
#Training using LinearRegression Model

from sklearn.linear_model import LinearRegression
lr_model=LinearRegression()
lr_model.fit(X_train,Y_train)
```

```python
#Testing
lr_pred=lr_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(Y_test,lr_pred)
print(MAE)
```

```python
from sklearn.metrics import r2_score
r2Score=r2_score(Y_test,lr_pred)
print(r2Score)
```

**MODEL II**

```python
from sklearn.tree import DecisionTreeRegressor
dt_model=DecisionTreeRegressor()
dt_model.fit(X_train,Y_train)
```

```python
dtpred=dt_model.predict(X_test)
```

```python
mae_dt=mean_absolute_error(Y_test,dtpred)
print(mae_dt)
```

```python
r2Score_dt=r2_score(Y_test,dtpred)
print(r2Score_dt)
```
