Analyzing Unemployment Rate

```python
#Import visualizing library
import matplotlib.pyplot as plt
import seaborn as sns

region=data2[data2['Region']== 'Andhra Pradesh']
plt.figure(figsize=(12, 12))
sns.lineplot(data=region, x=' Date',y=' Estimated Unemployment Rate (%)' , marker='o')
plt.title("unemployment rate in Andhra Predhesh over time")
plt.tight_layout()
```

```python
Average=data2.groupby('Region')[' Estimated Unemployment Rate (%)'].mean()
print(Average)
```

```python
Average.index
Average.values
Average.info()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x=Average.index, y=Average.values)
plt.title("Average unemployment rate based on region")
plt.xlabel('Region')
plt.ylabel('Unemployment_rate(%)')
plt.xticks(rotation=90)
plt.tight_layout()
```
