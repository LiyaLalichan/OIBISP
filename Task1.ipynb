```python
from sklearn.datasets import load_iris
```

```python
#Load Dataset
data=load_iris()
```

```python
#Exploring dataset
print(data.data.shape)
print(data.data)
print(data.target.shape)
print(data.target)
print(data.feature_names)
print(data.target_names)
```

```python
#Converting into a dataframe
import pandas as pd
#df uisng feature values and feature name
df=pd.DataFrame(data=data.data,columns=data.feature_names)
df['target'] = data.target_names[data.target]
```

```python
#Split data into training and testing
from sklearn.model_selection import train_test_split
X=data.data
Y=data.target
```

```python
from os import PRIO_PGRP
print(X)
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
print("Training feature set shape:", X_train.shape)
print("Testing feature set shape:", X_test.shape)
print("Training target set shape:", Y_train.shape)
print("Testing target set shape:", Y_test.shape)
```

```python
#Selecting and import model for training
from sklearn.ensemble import RandomForestClassifier
```

```python
#creating model
randf_model=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
#model fitting
randf_model.fit(X_train,Y_train)
```

```python
#Model tresting
model_prediction=randf_model.predict(X_test)
```

```python
#Model evaluation
from sklearn.metrics import accuracy_score,classification_report
```

```python
accuracy=accuracy_score(Y_test,model_prediction)
print("Accuracy of RNDOMFOREST MODEL",accuracy)
```

```python
print("Classifictaion report\n",classification_report(Y_test,model_prediction,target_names=data.target_names))
```

```python
#Model 2=Logistic regression
from sklearn.linear_model import LogisticRegression
model_logReg=LogisticRegression(max_iter=200,random_state=42)
model_logReg.fit(X_train,Y_train)
```

```python
prediction_model2= model_logReg.predict(X_test)
```

```python
accuracy_model2=accuracy_score(Y_test,prediction_model2)
print("Accuracy of model2",accuracy_model2)
```
